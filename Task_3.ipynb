```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe=pd.read_csv("SampleSuperstore.csv")
dataframe.shape
```

```python
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
dataframe
```

```python
df = dataframe.rename(columns={'Ship Mode': 'Ship_Mode','Sub-Category':'sub_category'})
```

```python
df
```

```python
df.Ship_Mode.value_counts()
```

```python
df.Segment.value_counts()
```

```python
df.Category.value_counts()
```

```python
df.sub_category.value_counts()
```

```python
import seaborn as sns
chart = sns.countplot(df.Ship_Mode)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
```

```python
df1 = df.groupby(['Ship_Mode'])['Profit'].sum().to_frame().reset_index()
```

```python
df1
```

```python
df2 = df.groupby(['City'])['Profit'].sum().to_frame().reset_index()
df2
```

```python
df3 = df.groupby(['sub_category'])['Profit'].sum().to_frame().reset_index()
df3
```

```python
df["cp"] = df['Sales']-df["Profit"]
```

```python
df4=df.groupby("sub_category")["cp"].sum()
```

```python
df4.plot(kind="bar")
```

```python
plt.figure(figsize = (100,200))
sns.catplot(x="sub_category", y="Profit",data=df3, kind="bar",height=6, aspect=.7)
plt.xticks(rotation = 90)
```

```python
group=df.groupby("sub_category")["Sales","Profit"].sum()
group.plot(kind="bar",width=0.8)

plt.show()
```

```python
sns.pairplot(df, hue = 'Ship_Mode')
```

```python
sns.pairplot(df,hue='Category')
```

```python
corr=df.corr()
sns.heatmap(corr,annot=True)
plt.show()
```

```python
plt.scatter(x = df["Sales"], y = df["Profit"])
```

```python
df['State'].value_counts()
```

```python
fig = plt.figure(figsize =(10, 20)) 

df['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

```python
fig = plt.figure(figsize =(30, 20)) 
sns.countplot(x="sub_category",data=df,hue="Region")
```

```python
state_wise=df.groupby("State")["Sales"].sum()
fig = plt.figure(figsize =(30, 20)) 
state_wise.plot(kind="bar")
```

```python
region_wise=df.groupby("Region")["Sales"].sum()
fig = plt.figure(figsize =(10, 7)) 
region_wise.plot(kind="bar")
```
